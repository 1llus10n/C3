```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct= ColumnTransformer([("on",OneHotEncoder(drop='first'),[1,4,5])],remainder='passthrough')
df=ct.fit_transform(df)
df
```

```python
from scipy import stats
```

```python
stats.zscore(df)
```

```python
z=np.abs(stats.zscore(df))
```

```python
z
```

```python
thres=3
np.where(z>thres)
```

```python
dfg=pd.DataFrame(df)
```

```python
dfg
```

```python
df_no_out=dfg[(z<thres).all(axis=1)]
df_no_out
```

```python
x=df_no_out.iloc[:,1:8].values
x
```

```python
y=df_no_out.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
sr=LinearRegression()
```

```python
sr.fit(x_train,y_train)
```

```python
y_pred=sr.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
a_train,a_test,b_train,b_test=train_test_split(x,y,test_size=0.2,random_state=0)
```
